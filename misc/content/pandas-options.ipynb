## Max Number of Columns to Display

```python
pd.options.display.max_columns = 100
```

## Max Number of Rows to Display

```python
pd.options.display.max_rows = 500
```

## Max Column Width to Display

`pd.options.display.max_colwidth`

```python
import pandas as pd
df = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'link': ['http://a_super_loooooooooooooooooooooooooooooooong_link.com', '', '', '', '']
})
df
```

```python
pd.options.display.max_colwidth = -1
# pd.set_option('display.max_colwidth', -1)
df
```

## Display Formatting of Floating Numbers

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randint(100, 10000, size=(4, 2)), columns=['a', 'b'])
df.head()
```

```python
df.dtypes
```

```python
pd.options.display.float_format = '{:,}'.format
```

```python
df
```
